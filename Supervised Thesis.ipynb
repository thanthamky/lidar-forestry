```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./tree_DBH50_metrics_spctrl.csv')
```

```python
data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
encoder = LabelEncoder()
y = encoder.fit_transform(data['SpeciesNam'])
X = data.iloc[:, 6:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
encoder.classes_
```

```python
encoder.inverse_transform([22, 27, 56, 58, 74, 84]).tolist()
```

```python
print(classification_report(y_test, y_pred, labels=[22, 27, 56, 58, 74, 84], target_names=['Choerospondias axillaris',
 'Dipterocarpus gracilis',
 'Machilus affsalicina',
 'Magnolia baillonii',
 'Prunus javanica',
 'Sloanea sigun']))
```
